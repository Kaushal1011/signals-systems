```python
import numpy as np
def main():
    A4_octave = [440]
    for _ in range(11):
        A4_octave.append(A4_octave[-1] * 2**(1 / 12))
    A4_octave = np.array(A4_octave)
    return np.array([A4_octave / 16,
                     A4_octave / 8,
                     A4_octave / 4,
                     A4_octave / 2,
                     A4_octave,
                     A4_octave * 2,
                     A4_octave * 4,
                     A4_octave * 8])
```

```python
from pprint import pprint
x = main()
freq = {}
for i in range(8):
    freq[i] = x[i]
pprint(freq)
```
