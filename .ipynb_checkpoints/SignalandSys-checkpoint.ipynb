```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
```

```python
t=np.linspace(0,8*3,44100*3)
```

```python
cost=(np.cos(2*np.pi*256*t)).reshape(44100*3,1)
```

```python
plt.figure(figsize=(20,6))
plt.grid()
plt.plot(cost[0:100],"b-")

plt.show()
```

```python
z=np.random.uniform(-0.5,0.5,44100*3).reshape(44100*3,1)
plt.figure(figsize=(20,6))
plt.grid()
plt.plot(z[0:100],"or--")
```

```python
wnoise=z+cost
```

```python
plt.figure(figsize=(20,6))
plt.plot(wnoise[0:100],"r-")
plt.plot(cost[0:100],"b--")
plt.grid()
```

```python
scaled = np.int16(wnoise/np.max(np.abs(wnoise)) * 32767)
scaled[0:50]
```

```python
write('test.wav', 44100, scaled)
```
